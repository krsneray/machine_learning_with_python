```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Veriler.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['cinsiyet'] = le.fit_transform(df['cinsiyet'])
df = df.drop(columns='ulke',axis=1)
df = df.dropna()
df
```

```python
x = df.drop(columns='cinsiyet',axis=1).values
y = df['cinsiyet'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,random_state=0,test_size=0.33)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression(random_state = 0)
logr.fit(xtrain,ytrain)
```

```python
ypred = logr.predict(xtest)
print(ypred)
print(ytest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ypred,ytest)
print(cm)
```
