## Gerekli Kütüphaneler

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Veri Setinin Eklenmesi

```python
df = pd.read_csv('Maaslar.csv')
```

```python
df
```

## Bağımlı Ve Bağımsız Değişkenlerin Belirlenmesi

```python
x = df.drop(columns=['maas','unvan'],axis=1).values
y = df['maas'].values
```

## Modelin Oluşturulması

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Tahmin Ve Sonuç Karşılaştırması

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x))
```

```python
# Görüldüğü gibi polinomal olan bir veri seti lineer modelle eğitildiğinde büyük sapmalar olmakta.
# (Veri seti y eksenine kayarken lineer model olduğu gibi devam ediyor.)
```

## Polinomal Regresyon

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg= PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
plt.scatter(x,y)
plt.plot(x,lin_reg2.predict(x_poly),color='red') #Lineer tahmini x_poly cinsinden yaparak polinomal regresyon tahmini yapmış olduk. 
```

```python
from sklearn.preprocessing import PolynomialFeatures
```
