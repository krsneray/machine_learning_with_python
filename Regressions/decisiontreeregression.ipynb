## Gerekli Kütüphaneler 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Veri Setinin Eklenmesi

```python
df = pd.read_csv('Maaslar.csv')
```

```python
df
```

## Bağımlı Ve Bağımsız Değişkenlerin Belirlenmesi

```python
x = df.drop(columns=['maas','unvan'],axis=1).values
y = df['maas'].values
```

## Modelin Oluşturulması (Karar Ağacı)

```python
 from sklearn.tree import DecisionTreeRegressor
```

```python
r_dt = DecisionTreeRegressor(random_state=0)
r_dt.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,r_dt.predict(x),color='red')
```
