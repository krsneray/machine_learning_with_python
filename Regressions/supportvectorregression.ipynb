## Gerekli Kütüphaneler

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Veri Setinin Eklenmesi

```python
df = pd.read_csv('Maaslar.csv')
```

```python
df
```

## Bağımlı Ve Bağımsız Değişkenlerin Belirlenmesi

```python
x = df.drop(axis=1,columns=['maas','unvan']).values
y = df['maas'].values
x = x.reshape(-1,1)
y = y.reshape(-1,1) #Scaler kullanabilmek için 2 boyutlu hale geldi.
```

## Modelin Oluşturulması

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
scaler2 = StandardScaler()
y_scale = scaler2.fit_transform(y)
```

## Destek Vektör Regresyonu

```python
from sklearn.svm import SVR
```

```python
svr_reg = SVR(kernel = 'rbf')
svr_reg.fit(x_scale,y_scale.ravel()) #Ravel kullanarak y_scale tekrar tek boyutlu hale geldi.
```

```python
plt.scatter(x_scale,y_scale)
plt.plot(x_scale,svr_reg.predict(x_scale),color='red')
```
